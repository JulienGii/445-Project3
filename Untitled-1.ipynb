```python
import p2

# Constant initialization
CORPUS_DIRECTORY = "Reuters-21578"
OUTPUT_DIRECTORY = "OutputsScript2/"
NUM_FILES_TO_RETRIEVE = 2

tokenized_doc = p2.pipeline(OUTPUT_DIRECTORY, CORPUS_DIRECTORY, NUM_FILES_TO_RETRIEVE)
```

```python
tokens_file1 = tokenized_doc[2] # one item per file
print(tokens_file1)
print(tokens_file1[0])
print(tokens_file1[1])
```

```python
# project2: module that, while there are still more documents to be processed, accepts a document as a list/stream of tokens and outputs term-documentID pairs to a list F
f = []
for docId, tokens in tokens_file1:
    for token in tokens:
        f.append((token, docId))

f = set(f) # remove duplicates
print(f)
print(len(f)) # number of pair (token, docId)

```

```python

sorted_pairs = sorted(f, key=lambda x: x[0])
print(sorted_pairs)
```

```python
inverted_index = []
postings = [] # posting list for a given token
previous_token = sorted_pairs[0][0] # initialize
for token, docId in sorted_pairs:


    # if this token is the same as the previous one, add the docId to posting list
    if token == previous_token:
        postings.append(docId)
        continue
    # this token is different than the last one, insert into posting the previous token with its posting
    else:
        inverted_index.append((previous_token, postings))
        previous_token = token
        postings = [docId]
print(inverted_index)
```

inverted_index is goooood

```python
for posting in inverted_index:
    print(posting)
```

```python
from p2 import process_query
query = 'Cleared toronto can\'t'
processed_q = process_query(query)[0][1]
print(processed_q)
```

```python
for token in processed_q:
    for term, posting in inverted_index:
        if token == term:
            print('found term ', term, 'in index')
            print(posting)
```
